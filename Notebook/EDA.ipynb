```python
fitness_data["Month"] = fitness_data["BookingEndDate"].dt.month_name()
```

```python
fitness_data["Hour"] = fitness_data["BookingStartTime"].dt.hour
```

```python
fitness_data["DayName"] = fitness_data["BookingEndDate"].dt.day_name()
```

```python
fitness_data["Session"] = fitness_data["Hour"].apply(
    lambda x:
    "Morning" if x < 12
    else "Afternoon" if x < 17
    else "Evening"
)
```

```python
fitness_data["Booking_Rate"] = (
    fitness_data["Number Booked"] /
    fitness_data["MaxBookees"]
) * 100
```

```python
fitness_data["Revenue"] = (
    fitness_data["Price (INR)"] *
    fitness_data["Number Booked"]
)
```

```python
activity_bookings = (
    fitness_data.groupby("ActivityDescription")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=activity_bookings.values,
    y=activity_bookings.index
)

# Add data labels
for i in ax.containers:
    ax.bar_label(i, fmt='%d', padding=1)

plt.title("Top 10 Fitness Activities by Total Bookings")
plt.xlabel("Total Bookings")
plt.ylabel("Fitness Activity")

plt.tight_layout()
plt.show()
```

```python
top10_revenue = (
    fitness_data
    .groupby("ActivityDescription")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10))
top10_revenue.plot(
    kind="bar",
    figsize=(12,6)
    
)

plt.title("Top 10 Activities by Revenue")
plt.xlabel("Activity")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    fitness_data["MaxBookees"],

    fitness_data["Number Booked"]
)

plt.title("Class Capacity vs Bookings")
plt.xlabel("Maximum Booking")
plt.ylabel("Bookings")

plt.grid(alpha=0.3)

plt.show()
```

```python
month_order = ["April", "May", "June"]

month_booking = (
    fitness_data.groupby("Month")["Number Booked"]
    .sum()
    .reindex(month_order)
)

plt.figure(figsize=(6,4))

ax = month_booking.plot(kind="bar")

plt.title("Total Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Total Bookings")

plt.xticks(rotation=0)
plt.grid(axis="y", alpha=0.3)

# Data Labels
for i, value in enumerate(month_booking):
    ax.text(i, value, str(int(value)),
            ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
month_booking = (
    fitness_data.groupby("DayName")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(6,4))

ax = month_booking.plot(kind="bar")

plt.title("Total Bookings by Day")
plt.xlabel("Day")
plt.ylabel("Total Bookings")
plt.xticks(rotation=25)

# Data Labels
for i, value in enumerate(month_booking):
    ax.text(i, value, str(int(value)),
            ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
site_bookings = (
    fitness_data.groupby("Session")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
  
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    
    x=site_bookings.index,
    y=site_bookings.values,
)

# Add data labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=1)

plt.title("Total Bookings by Session")
plt.xlabel("Session")
plt.ylabel("Total Bookings")

plt.tight_layout()
plt.show()
```

```python
session_revenue = (
    fitness_data
    .groupby("Session")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

session_revenue
session_revenue.plot(
    kind="bar",
    figsize=(7,5),
    color=["skyblue","orange","green"]
)

plt.title("Revenue by Session")
plt.xlabel("Session")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=0)
plt.show()
```

```python
site_bookings = (
    fitness_data.groupby("ActivitySiteID")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
  
)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    # x=site_bookings.values,
    x=site_bookings.index,
    y=site_bookings.values,
)

# Add data labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=1)

plt.title("Activity Sites by Total Bookings")
# plt.xlabel("Total Bookings")
plt.xlabel("Activity Site"),
plt.ylabel("Total Bookings")
plt.tight_layout()
plt.show()
```

```python
avg_price = (
    fitness_data
    .groupby("ActivitySiteID")["Price (INR)"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

avg_price.plot(kind="bar")

plt.title("Average Class Price by Site")
plt.xlabel("Site")
plt.ylabel("Average Price")

plt.xticks(rotation=0)

plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
location_revenue = (
    fitness_data
    .groupby("ActivitySiteID")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

location_revenue
location_revenue.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Revenue by Site")
plt.xlabel("Site")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=0)
plt.show()
```
