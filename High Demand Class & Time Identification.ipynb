## High Demand Class and Time Identification

### Import necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load the Dataset


```python
fitness_data = pd.read_excel("Cleaned_Preprocessed_Dataset_Yogeshwar Tribhuvan.xlsx")
```

### Dataset Information

```python
fitness_data.info()
fitness_data.head()
```

### Data Preparation

```python
fitness_data["BookingEndDate"] = pd.to_datetime(fitness_data["BookingEndDate"])

fitness_data["DayName"] = fitness_data["BookingEndDate"].dt.day_name()

fitness_data["Month"] = fitness_data["BookingEndDate"].dt.month_name()

fitness_data["Hour"] = pd.to_datetime(
    fitness_data["BookingStartTime"].astype(str)
).dt.hour
```

```python
def session(hour):
    if hour < 12:
        return "Morning"
    elif hour < 17:
        return "Afternoon"
    else:
        return "Evening"

fitness_data["Session"] = fitness_data["Hour"].apply(session)
```

### High Demand Classes

```python
class_demand = (
    fitness_data.groupby("ActivityDescription")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

class_demand
```

### High Demand Locations

```python
location_demand = (
    fitness_data.groupby("ActivitySiteID")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

location_demand
```

### High Demand Days

```python
day_demand = (
    fitness_data.groupby("DayName")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

day_demand
```

### High Demand Sessions

```python
session_demand = (
    fitness_data.groupby("Session")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

session_demand
```

### Monthly Demand Analysis

```python
month_demand = (
    fitness_data.groupby("Month")["Number Booked"]
    .sum()
    .sort_values(ascending=False)
)

month_demand
```

### Correlation Analysis

```python
correlation = fitness_data[
    ["Price (INR)", "MaxBookees", "Number Booked"]
].corr()

correlation
```

```python
import joblib
high_demand = {
    "Top Class": class_demand.idxmax(),
    "Top Location": location_demand.idxmax(),
    "Top Day": day_demand.idxmax(),
    "Top Session": session_demand.idxmax()
}

joblib.dump(high_demand, "high_demand.pkl")

print("High Demand insights saved!")
```
