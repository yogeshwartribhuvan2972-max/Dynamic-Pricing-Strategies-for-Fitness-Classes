### Step 5: Rename Columns

```python
april_may.rename(
    columns={"BookingEndDateTime (Month / Day / Year)": "BookingEndDate"},
    inplace=True
)

june.rename(
    columns={"BookingEndDateTime (Month / Day / Year)": "BookingEndDate"},
    inplace=True
)
```

### Task: Data Integration, Merging & Quality Assurance

### Step 1: Understand the Data Integration Requirement

```python
fitness_data = pd.concat(
    [april_may, june],
    ignore_index=True
)
```

### Quality Report
The merged dataset was successfully validated after integration.

The April–May and June datasets were successfully combined into a single dataset.
Data types remained consistent after integration.
No negative values were found in the Price (INR), MaxBookees, or Number Booked columns.
A few records were found where the Number Booked exceeded the MaxBookees. These records were retained because there was insufficient evidence to determine whether the booking count or the maximum capacity was incorrect.
The integrated dataset passed the required quality checks and is ready for Exploratory Data Analysis (EDA).
