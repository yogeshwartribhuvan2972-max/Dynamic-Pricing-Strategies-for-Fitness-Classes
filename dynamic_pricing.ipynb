### Step 5: Rename Columns

```python
# Rename the Booking End Date column for better readability

april_may.rename(
    columns={"BookingEndDateTime (Month / Day / Year)": "BookingEndDate"},
    inplace=True
)

june.rename(
    columns={"BookingEndDateTime (Month / Day / Year)": "BookingEndDate"},
    inplace=True
)
```

### Step 6: Handle Missing Values

```python
#display only null value data
april_may[april_may["Price (INR)"].isnull()]
```

```python
# Display Body Cond classes to determine the correct missing price
april_may[
    april_may["ActivityDescription"].str.contains("Body Cond", case=False)
]
```

```python
# Display Combat Aero classes to determine the correct missing price
april_may[
    april_may["ActivityDescription"].str.contains("Combat Aero", case=False)
]
```

```python
# Display june Combat Aero classes to determine the correct missing price because april and may missing price
june[
    june["ActivityDescription"].str.contains("Combat Aero", case=False)
]
```

```python
#fill missing values
april_may.loc[
    (april_may["ActivityDescription"].str.contains("Body Cond", case=False)) &
    (april_may["Price (INR)"].isnull()),
    "Price (INR)"
] = 1299
```

```python
#fill missing values
april_may.loc[
    (april_may["ActivityDescription"].str.contains("Combat Aero", case=False)) &
    (april_may["Price (INR)"].isnull()),
    "Price (INR)"
] = 1499
```

```python
#recheck null value
april_may["Price (INR)"].isnull().sum()
```

```python
# Convert the Booking End Date column to datetime format for both datasets
april_may["BookingEndDate"] = pd.to_datetime(
    april_may["BookingEndDate"],
    format="%d-%b-%y"
)
june["BookingEndDate"] = pd.to_datetime(
   june["BookingEndDate"],
    format="%d-%b-%y"
)
```

```python
# Convert the Booking Start Time column to datetime format for both datasets
april_may["BookingStartTime"] = pd.to_datetime(
    april_may["BookingStartTime"],
    format="%H:%M:%S"
)
june["BookingStartTime"] = pd.to_datetime(
    june["BookingStartTime"],
    format="%H:%M:%S"
)
```
